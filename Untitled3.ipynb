```python
import matplotlib.pyplot as plt
```

```python
VIS2_MEAN = 214.7
AUD2_MEAN = 424.4
VIS2_STD = 31.08
AUD2_STD = 35.87
```

```python
# build a bar plot
plt.bar(['Person 2 Visual', 'Person 2 Auditory'],[VIS2_MEAN, AUD2_MEAN], yerr=[VIS2_STD, AUD2_STD], capsize=10)
plt.ylabel('Speed (msec)')
plt.xlabel("Reflex Type")
plt.title('Reflex Speed, Visual vs. Auditory')
```
